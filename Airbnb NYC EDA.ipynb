### 4.DATA CLEANING

```python
df.columns            # Checking for all columns.
```

```python
df.dtypes          # Understanding data type of a column.
```

**Observation:**checking for duplicate features and understanding datatype of a features.

```python
df.nunique()            # No. of unique values in column wise.
```

**Observation:** Checking for no.of unique values in feature wise.

```python
len(df[df.duplicated()])    # Checking for duplicate rows. If we have duplicate rows then drop duplicates using df.drop_duplicates().
```

```python
df.dropna(how='all',axis=1,inplace=True)         # If missing value is more than 25% then drop the column.
```

```python
df.shape         # After dropping null values the shape of the dataset.
```

**Observation:** checking for duplicate rows and dropping them.As there are no duplicates we are not dropping.If the missing values has 25% above value then drop the column.

```python
df.isnull().sum()      # Counting no of null values in each column wise.
```

```python
(df.isnull().sum()/df.shape[0])*100            # Counting no. of null values in each column wise.``
```

**Observation:** Counting no. of null values in each feature wise.

```python
df["name"].fillna(df["name"].mode()[0],inplace=True)          # If the column is categorical fill the missing values with mode()[0].
df["host_name"].fillna(df["host_name"].mode()[0],inplace=True)
```

```python
df["last_review"].fillna(df["last_review"].median(),inplace=True)       # If the column is numerical and if it has any outliers fill the missing values with median.
```

```python
df["reviews_per_month"].fillna(df["reviews_per_month"].median(),inplace=True)   # If the column is numerical and if it  has any outliers fill the missing values with mean.
```

**Observation:** If the feature is categorical then fill the missing values with mode()[0]i.e.., univariate.If the feature is numerical and if it has any outliers then fill the missing values with median otherwise with mean.

```python
df.isnull().sum()           # After filling missing values checking whether null values are there in column wise.
```

**Deriving New Columns**

```python
df['last_review_month'] = df['last_review'].dt.month_name().str[:3]
df['last_review_year'] = pd.DatetimeIndex(df['last_review']).year
```

**Observation:** Extracting month and year from the last_review column and deriving with new columns.

### 6.REMOVING OUTLIERS

checking outliers.

```python
plt.figure(figsize=(15,8))
print("Before removing Outliers:",sns.boxplot(df))
plt.xticks(rotation=45)
plt.show()
```

```python
df.columns
```

```python
Q1=df['minimum_nights'].quantile(0.25)
Q3=df['minimum_nights'].quantile(0.75)
IQR=Q3-Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
print(IQR)
print(lower_bound)
print(upper_bound)
```

```python
df = df[(df['minimum_nights']>lower_bound)&(df['minimum_nights']<upper_bound)]
```
